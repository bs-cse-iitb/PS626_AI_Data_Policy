```python
# exponential time complexity
def fibonacci1(n):
    if n <= 1:
        return 1   
    return fibonacci(n - 1) + fibonacci(n - 2)


def fibonacci(n, intermediate_dict={}):
    #print(intermediate_dict)
    if n <= 1:
        return 1 #n if start with zero
    if n not in intermediate_dict:
        intermediate_dict[n] = fibonacci(n-1, intermediate_dict) + fibonacci(n-2,intermediate_dict)
    return intermediate_dict[n]

no_of_elements = 50
# generally fibonacci number start with 0 1 1 2, but in questions it start with 1 1 2  
print("Fibonacci series:")
for i in range(no_of_elements):
    print("Element",i+1,"--------------",fibonacci(i))
```

```python
import random
#insertion sort
def ascending_order(list1):
    size = len(list1)
    for i in range(1, size): 
        key = list1[i]
        j = i-1
        while j >= 0 and key < list1[j] :
                list1[j+1] = list1[j]
                j = j - 1
        list1[j+1] = key        
    return list1

random.seed(0)
no_of_element=50
numbers = [random.randint(1,no_of_element) for i in range(no_of_element)]
print("Unsorted number:",numbers)

sorted_numbers = ascending_order(numbers)
print("Sorted numbers in ascending order:", sorted_numbers)
```
