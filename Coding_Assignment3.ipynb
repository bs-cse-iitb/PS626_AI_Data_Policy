```python
import os
# Create a directory to store the text files
if not os.path.exists('txt_files'):
    os.mkdir('txt_files')

# Loop through the docs list and create a text file for each document
for i, doc in enumerate(docs):
    # Create a new file with the name PIB+i.txt
    filename = f"Research_Article_{i+1}.txt"
    filepath = os.path.join('txt_files', filename)
    
    # Write the content of the document to the text file
    with open(filepath, 'w') as f:
      f.write(doc)
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english")
from bertopic.vectorizers import ClassTfidfTransformer

# ctfidf_model = ClassTfidfTransformer(reduce_frequent_words=True)
# topic_model = BERTopic(vectorizer_model=vectorizer_model,ctfidf_model=ctfidf_model,language="english",verbose=True,min_topic_size=2,top_n_words=8)#50
topic_model = BERTopic(vectorizer_model=vectorizer_model,language="english",verbose=True,min_topic_size=2,top_n_words=100)#50(8/3)
topics, probs = topic_model.fit_transform(docs)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def create_wordcloud(model, topic):
    text = {word: value for word, value in model.get_topic(topic)}
    wc = WordCloud(background_color="white", max_words=1000)
    wc.generate_from_frequencies(text)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()

# Show wordcloud
for i in range():
  create_wordcloud(topic_model, topic=i)
```
